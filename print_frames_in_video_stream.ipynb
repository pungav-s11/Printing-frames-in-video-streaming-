```python
import cv2
from google.colab.patches import cv2_imshow

def process_video(video_path, interval_seconds):
    # Capture the video from the file
    cap = cv2.VideoCapture(video_path)

    # Get the frame rate of the video
    frame_rate = cap.get(cv2.CAP_PROP_FPS)
    print(f"Frame Rate: {frame_rate} frames per second")

    # Calculate the interval in terms of frames
    interval = int(frame_rate * interval_seconds)

    # Frame count
    frame_count = 0

    while cap.isOpened():
        ret, frame = cap.read()
        if not ret:
            break

        if frame_count % interval == 0:
            print(f"Displaying frame at {frame_count // frame_rate} seconds")
            cv2_imshow(frame)
            if cv2.waitKey(0) & 0xFF == ord('q'):
                break

        frame_count += 1

    cap.release()
    cv2.destroyAllWindows()

# Get the video file path and interval from the user
video_path = input("Enter the path to your video file: ")
interval_seconds = int(input("Enter the interval (in seconds) between frames to display: "))
process_video(video_path, interval_seconds)
```
